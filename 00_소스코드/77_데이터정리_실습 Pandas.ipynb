```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
news = {
    '날짜': [],
    '기사개수': []
}

for i in range(1, 31):
    
    news_count = 0
    start = 1
    
    while True:
        
        day = ''
        
        if i < 10:
            day = '0' + str(i)
        else:
            day = str(i)
        
        params = {
            'where': 'news',
            'sm': 'tab_pge',
            'query': '삼성전자',
            'sort': 1,
            'photo': 0,
            'field': 0,
            'pd': 3,
            'ds': '2021.10.' + day,
            'de': '2021.10.' + day,
            'mynews': 0,
            'office_type': 0,
            'office_section_code': 0,
            'news_office_checked': '',
            'nso': f'so:dd,p:from202111{day}to202111{day},a:all',
            'start': start
        }
        response = requests.get('https://search.naver.com/search.naver', params=params)
        soup = BeautifulSoup(response.text, 'html.parser')

        news_count += len(soup.select('ul.list_news > li'))

        if soup.select_one('div.not_found02') != None:
            print("기사끝")
            break

        start += 10
    
    news['날짜'].append('202110' + day)
    news['기사개수'].append(news_count)
    print('202110' + day)
```

```python
news_df = pd.DataFrame(news, columns=['기사개수'], index=news['날짜'])
```

```python
stock = {
    '날짜': [],
    '코스피지수': []
}

headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.45 Safari/537.36'
}

for page in range(3,5):
    response = requests.get('https://finance.naver.com/item/sise_day.naver?code=005930&page=' + str(page), headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')

    for item in soup.select('table.type2 tr'):
        if len(item.select('td'))== 7:
            stock['날짜'].append(item.select('td')[0].text.replace('.',''))
            stock['코스피지수'].append(item.select('td')[1].text.replace(',',''))
```

```python
stock_df = pd.DataFrame(stock, columns=['코스피지수'], index=stock['날짜'])
```
